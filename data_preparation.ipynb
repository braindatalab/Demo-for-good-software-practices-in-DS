```python
edges = np.geomspace(100001, 4111111766, 17)
edges
```

```python
bin_edges = [(edges[x], edges[x+1]) for x in range(16)]
```

```python
probs
```

```python
len(probs.keys())
```

```python
sorted(probs.keys())
```

```python
bin_probs = dict()
for idx, bin_edge in enumerate(bin_edges):
    start_prob_idx = max(0, idx-1)
    stop_prob_idx = min(len(sorted(probs.keys()))-1, idx+1)
    print(start_prob_idx, stop_prob_idx)
    bin_prob_dict = dict()
    for i in range(start_prob_idx, stop_prob_idx+1):
        bin_prob_dict[sorted(probs.keys())[i]] = probs[sorted(probs.keys())[i]]
    bin_probs[bin_edge] = bin_prob_dict
        
```

```python
bin_probs
```

```python
probs.values()
```

```python
np.asarray(list(list(bin_probs.values())[0].values()))
```

```python
def draw_bathrooms(price):
    for bin, prob in bin_probs.items():
        low, high = bin
        if low <= price <= high: 
            prob_dist = np.array(list(prob.values()))
            return np.random.default_rng().choice(list(prob.keys()), p=prob_dist/prob_dist.sum())
    print("Outside the range")
```

```python
df4["BATHROOM_COUNT"] = df4["SALE_PRICE"].map(draw_bathrooms)
```
